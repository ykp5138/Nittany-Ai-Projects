```python
import tensorflow as tf
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler

pretrained_model = tf.keras.applications.ResNet50(
    include_top=False,
    input_shape=(150, 150, 3),
    pooling='avg',
    classes=2,
    weights='imagenet'
)

for layer in pretrained_model.layers:
    layer.trainable = False

x = pretrained_model.output
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(2, activation='softmax')(x)

model = tf.keras.models.Model(inputs=pretrained_model.input, outputs=x)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
import numpy as np
import tensorflow as tf

training_data = np.random.rand(100, 150, 150, 3)
training_labels = np.random.randint(0, 2, size=(100,))

training_labels = tf.keras.utils.to_categorical(training_labels, num_classes=2)

pretrained_model = tf.keras.applications.ResNet50(
    include_top=False,
    input_shape=(150, 150, 3),
    pooling='avg',
    classes=2,
    weights='imagenet'
)
num=0
for layer in pretrained_model.layers:
    num+=1
    layer.trainable = False
print(num)

x = pretrained_model.output
x = tf.keras.layers.Flatten()(x)

x = tf.keras.layers.Dense(2, activation='softmax')(x)

model = tf.keras.models.Model(inputs=pretrained_model.input, outputs=x)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(training_data, training_labels, epochs=20)

history = model.fit(training_data, training_labels, epochs=20)
avg_accuracy = np.mean(history.history['accuracy'])
print(f"Average accuracy: {avg_accuracy}")
```
